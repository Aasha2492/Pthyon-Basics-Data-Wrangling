```python
# Task 2: Store marks in 5 subjects in a list and print the average

subjects = ["Math", "English", "Physics", "Chemistry", "Computer"]
marks    = [78, 85, 90, 72, 88]

print("Subject-wise marks:")
for subject, mark in zip(subjects, marks):
    print(f"  {subject}: {mark}")

average = sum(marks) / len(marks)
print(f"\nAverage Marks: {average:.2f}")
```

```python
# Task 3: Dictionary of 5 students – print who passed and who failed

students = {
    "Ali":    72,
    "Sara":   45,
    "Bilal":  58,
    "Fatima": 38,
    "Usman":  91
}

print("Student Results:")
print("-" * 30)
for name, score in students.items():
    result = "✅ Passed" if score >= 50 else "❌ Failed"
    print(f"  {name:<10} Marks: {score}  →  {result}")
```

```python
# Task 4 & 5: grade_calculator(marks) – returns grade based on average

def grade_calculator(marks):
    """
    Takes a list of marks and returns a grade:
      A    → average >= 80
      B    → average >= 65
      C    → average >= 50
      Fail → average <  50
    """
    if not marks:
        return "No marks provided"

    avg = sum(marks) / len(marks)

    if avg >= 80:
        grade = "A"
    elif avg >= 65:
        grade = "B"
    elif avg >= 50:
        grade = "C"
    else:
        grade = "Fail"

    return grade


# --- Test the function ---
test_cases = [
    [90, 85, 88, 92, 78],   # Should be A
    [70, 68, 72, 65, 60],   # Should be B
    [55, 52, 58, 50, 53],   # Should be C
    [40, 35, 45, 30, 42],   # Should be Fail
]

print("Grade Calculator Test Results:")
print("-" * 40)
for marks_list in test_cases:
    avg   = sum(marks_list) / len(marks_list)
    grade = grade_calculator(marks_list)
    print(f"  Marks: {marks_list}")
    print(f"  Average: {avg:.1f}  →  Grade: {grade}")
    print()
```

```python
df = df_raw.copy()

# ── 1. Standardize text case ────────────────────────────────────────────────
df['City']    = df['City'].str.strip().str.title()       # e.g. KARACHI → Karachi
df['Product'] = df['Product'].str.strip().str.title()    # e.g. mouse   → Mouse
print("✅ 1. Standardized City and Product text case")

# ── 2. Fix Quantity: convert non-numeric → NaN, then handle ─────────────────
df['Quantity'] = pd.to_numeric(df['Quantity'], errors='coerce')
print("✅ 2. Converted non-numeric Quantity values to NaN")

# ── 3. Fix negative Quantity ─────────────────────────────────────────────────
df.loc[df['Quantity'] < 0, 'Quantity'] = np.nan
print("✅ 3. Set negative Quantity values to NaN")

# ── 4. Fix Price: ensure numeric ─────────────────────────────────────────────
df['Price'] = pd.to_numeric(df['Price'], errors='coerce')
print("✅ 4. Ensured Price column is numeric")

# ── 5. Standardize OrderDate format ──────────────────────────────────────────
df['OrderDate'] = pd.to_datetime(df['OrderDate'], infer_datetime_format=True, errors='coerce')
print("✅ 5. Standardized OrderDate to datetime format")

# ── 6. Drop rows with missing critical fields (CustomerName or Product) ──────
before = len(df)
df.dropna(subset=['CustomerName', 'Product'], inplace=True)
print(f"✅ 6. Dropped {before - len(df)} row(s) with missing CustomerName or Product")

# ── 7. Fill remaining missing values ─────────────────────────────────────────
#   Price  → fill with median price of that Product
df['Price'] = df.groupby('Product')['Price'].transform(
    lambda x: x.fillna(x.median())
)
#   Quantity → fill with 1 (minimum reasonable order)
df['Quantity'].fillna(1, inplace=True)
#   City → fill with 'Unknown'
df['City'].fillna('Unknown', inplace=True)
#   OrderDate → fill with median date
df['OrderDate'].fillna(df['OrderDate'].median(), inplace=True)
print("✅ 7. Filled remaining missing values")

# ── 8. Remove duplicate rows ──────────────────────────────────────────────────
before = len(df)
df.drop_duplicates(inplace=True)
print(f"✅ 8. Removed {before - len(df)} duplicate row(s)")

# ── 9. Correct data types ─────────────────────────────────────────────────────
df['OrderID']  = df['OrderID'].astype(int)
df['Quantity'] = df['Quantity'].astype(int)
df['Price']    = df['Price'].astype(float)
print("✅ 9. Corrected all column data types")

# ── 10. Reset index ───────────────────────────────────────────────────────────
df.reset_index(drop=True, inplace=True)

print("\n✅ Data cleaning complete!")
```
